```python
import nltk
from nltk.corpus import stopwords, wordnet
from nltk.tokenize import word_tokenize
import string

# Download NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Sample journal entries
journal_entries = [
    "Today, I am grateful for the love and support of my family.",
    "I appreciate the little things in life, like the warmth of the sun and the beauty of nature.",
    "Expressing gratitude for the kindness of strangers fills my heart with joy.",
    "Today was a productive day! Started the morning with a brisk walk, which helped clear my mind. At work, I tackled a challenging project and made significant progress. Took a break to have lunch with a friend, which was refreshing. In the evening, I attended a yoga class, followed by some reading before bed.",
    "Woke up feeling exhausted today, probably due to a restless night's sleep. Despite that, managed to push through and complete my tasks at work. Took some time in the evening to relax with a hot bath and a good book. Looking forward to getting a better night's sleep tonight.",
    "Today was a rollercoaster of emotions. Dealt with some unexpected challenges at work that left me feeling stressed and overwhelmed. Took a long walk in the park during lunch to clear my head, which helped a bit. Spent the evening journaling and reflecting on how to better manage stress in the future.",
    "Had a fun and laid-back day today. Met up with friends for brunch and spent the afternoon exploring a new neighborhood. Enjoyed some downtime in the evening, watching a movie and indulging in some self-care rituals. Grateful for days like these that remind me to slow down and appreciate the little moments.",
    "Feeling under the weather today, so I decided to take it easy. Worked from home in the morning to avoid spreading any germs. Spent the afternoon resting, drinking tea, and catching up on a few episodes of my favorite TV show. Hoping to feel better tomorrow.",
    "Today was filled with exciting opportunities! Had a productive meeting with a potential client that went really well. Celebrated with dinner at my favorite restaurant. Feeling motivated and inspired to keep pushing forward with my goals.",
    "Battled with procrastination today and struggled to stay focused on my tasks. Ended up feeling disappointed in myself for not being more productive. However, I'm trying to practice self-compassion and remind myself that everyone has off days. Tomorrow is a new day to start fresh.",
    "Spent the day volunteering at a local shelter, and it was incredibly rewarding. Seeing the impact I can make in my community fills my heart with joy. Grateful for the opportunity to give back and make a difference in someone else's life.",
    "Had a heart-to-heart conversation with a loved one today, and it brought us closer together. Communication is so important in maintaining healthy relationships, and I'm glad we were able to address some lingering issues. Feeling grateful for their understanding and support.",
    "i have facing very much difficulty in my life right now and have so many problems to tackle in the life",
    "Took a spontaneous road trip with friends today, and it was just what I needed. Exploring new places and making memories with loved ones is food for the soul. Ended the day with a bonfire and stargazing, feeling content and at peace with the world.",
    # Add more journal entries here
]

# Core values keywords
core_values_keywords = {
    "Gratitude": {"grateful", "joy", "kindness", "thankful", "appreciation", "gratitude", "blessed", "contentment", "happiness", "appreciation", "thankfulness", "thankful", "grace", "joy", "kind-hearted", "generosity", "warmth", "positivity", "appreciation", "kindness", "heartfelt", "acknowledgment", "gratefulness"},
    "Humility": ["divine", "receptivity", "humble", "nothing", "modesty", "humility", "humbleness", "meekness", "modest", "unassuming", "self-effacing", "lowliness", "modesty", "simplicity", "humiliate", "unpretentiousness", "modest", "deference", "submissive", "down-to-earth", "respect", "reverence", "egoless", "unassuming"],
    "Perseverance": ["persistence", "endurance", "commitment", "dedication", "persistence", "perseverance", "endurance", "determination", "steadfastness", "resilience", "tenacity", "dedication", "commitment", "resolve", "grit", "persistence", "endurance", "staying power", "fortitude", "perseverance", "patience", "resilience", "diligence"],
    "Aspiration": ["highest", "goal", "ambition", "dream", "ambition", "aspiration", "goal", "dream", "desire", "aim", "objective", "purpose", "target", "longing", "yearning", "hope", "wish", "ambition", "aspiration", "goal-setting", "determination", "vision", "dream"],
    "Receptivity": ["receive", "accept", "openness", "assent", "openness", "receptivity", "acceptance", "openness", "responsiveness", "willingness", "openness", "receptiveness", "openness", "willingness", "acceptance", "responsiveness", "open-mindedness", "flexibility", "receptivity", "willingness", "understanding", "embracing", "allowance"],
    "Progress": ["growth", "development", "evolution", "improvement", "growth", "development", "progress", "improvement", "advancement", "evolution", "growth", "development", "progress", "betterment", "advancement", "enhancement", "evolution", "maturation", "refinement", "progression", "innovation", "change", "evolution"],
    "Courage": ["risk", "bravery", "fearlessness", "boldness", "bravery", "courage", "fearlessness", "boldness", "valor", "bravery", "courage", "fearlessness", "boldness", "bravery", "daring", "intrepidity", "valor", "pluck", "nerve", "guts", "fortitude", "audacity", "valor", "courage"],
    "Goodness": ["kindness", "compassion", "selflessness", "altruism", "kindness", "compassion", "selflessness", "altruism", "goodness", "kindness", "compassion", "selflessness", "altruism", "benevolence", "empathy", "sympathy", "generosity", "humanity", "goodness", "goodwill", "kindness", "compassion", "benevolence", "caring"],
    "Generosity": ["giving", "selflessness", "liberation", "joy", "giving", "generosity", "selflessness", "liberation", "benevolence", "giving", "generosity", "selflessness", "liberation", "altruism", "magnanimity", "munificence", "liberality", "largesse", "charity", "benevolence", "giving", "open-handedness", "selflessness", "generosity"],
    "Equality": ["peace", "calm", "equilibrium", "harmony", "peace", "calm", "equilibrium", "harmony", "equality", "peace", "calm", "equilibrium", "harmony", "tranquility", "serenity", "balance", "peace", "tranquility", "serenity", "calmness", "equanimity", "balance", "stability", "peace", "harmony"],
    "Peace":["tranquility", "serenity", "harmony", "balance", "tranquility", "serenity", "harmony", "balance", "peace", "tranquility", "serenity", "harmony", "balance", "calm", "stillness", "peace", "quiet", "rest", "relaxation", "calmness", "serenity", "tranquility", "harmony", "equilibrium"],
    # Add more core values and their keywords here
}

# Function to preprocess a single journal entry
def preprocess_entry(entry):
    # Tokenization
    tokens = word_tokenize(entry)
    
    # Lowercasing
    tokens = [token.lower() for token in tokens]
    
    # Remove stopwords and punctuation
    stop_words = set(stopwords.words('english'))
    tokens = [token for token in tokens if token not in stop_words and token not in string.punctuation]
    
    return tokens

# Preprocess all journal entries
preprocessed_entries = [preprocess_entry(entry) for entry in journal_entries]

# Function to get synonyms of a word using WordNet
def get_synonyms(word):
    synonyms = set()
    for syn in wordnet.synsets(word):
        for lemma in syn.lemmas():
            synonyms.add(lemma.name())
    return synonyms

# Function to assign core value to preprocessed journal entry
def assign_core_value(entry, core_values_keywords):
    # Initialize dictionary to store keyword frequency for each core value
    keyword_freq = {core_value: 0 for core_value in core_values_keywords.keys()}
    
    # Count frequency of core value keywords and their synonyms in the entry
    for word in entry:
        for core_value, keywords in core_values_keywords.items():
            if word in keywords:
                keyword_freq[core_value] += 1
            else:
                # Check if word has any synonyms in core value keywords
                synonyms = get_synonyms(word)
                for syn in synonyms:
                    if syn in keywords:
                        keyword_freq[core_value] += 1
    
    # Get core value with maximum keyword frequency
    max_core_value = max(keyword_freq, key=keyword_freq.get)
    
    return max_core_value

# Assign core values to preprocessed entries
assigned_core_values = [assign_core_value(entry, core_values_keywords) for entry in preprocessed_entries]

# Count occurrences of each core value
core_value_counts = {core_value: assigned_core_values.count(core_value) for core_value in core_values_keywords}

# Calculate percentage for each core value
total_entries = len(journal_entries)
core_value_percentages = {core_value: (count / total_entries) * 100 for core_value, count in core_value_counts.items()}

# Print assigned core values for each entry
for i, core_value in enumerate(assigned_core_values, 1):
    print(f"Preprocessed Journal Entry {i}: {journal_entries[i-1]}")
    print(f"Assigned Core Value: {core_value}")
    print()

# Print core value percentages
for core_value, percentage in core_value_percentages.items():
    print(f"{core_value}: {percentage:.2f}%")
```
